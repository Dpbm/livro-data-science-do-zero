# Probability

```python
def probability(case, total):
    return case/total
```

```python
def independent_probability(E, F):
    return E*F
```

```python
# Coin flip: 2 heads
prob1 = probability(1, 2)
prob2 = probability(1, 2)
independent_probability(prob1, prob2)
```

```python
def dependent_probability(E, F):
    return 0 if F == 0 else E/F
```

```python
# probability of the 2 second child be a boy, since the first one is girl
prob1 = probability(1, 2)
prob2 = probability(1, 2)
dependent_probability(prob1, prob2)
```

| first | second |
|------|--------|
| boy  |  boy   |
| boy  |  girl  |
| girl |  boy   |
| girl |  girl  |

since the first is a girl, there's just two possibilities

| first | second |
|------|--------|
| girl |  boy   |
| girl |  girl  |

and since that there's just a way to have a boy after a girl, so the probability is:

$Prob(E|F) = 1/2$


```python
# both girls since, at least, one is a girl
prob1 = probability(1, 4)
prob2 = probability(3, 4)
dependent_probability(prob1, prob2)
```

| first | second |
|------|--------|
| boy  |  boy   |
| boy  |  girl  |
| girl |  boy   |
| girl |  girl  |

Once we know that at least one child is a girl, so we have three possibilities at total

| first | second |
|------|--------|
| boy  |  girl  |
| girl |  boy   |
| girl |  girl  |

so the probalitity of having two girls is: $Prob(E|F) = 1/3$

# Bayes Theorem!!!!

$P(E|F) = { P(E,F) \over P(F) }$

$P(F|E) = { P(E,F) \over P(E) }$

$ P(E|F)P(F) = P(F|E)P(E)$\
$ P(E|F) = {P(F|E)P(E) \over P(F)}$

# Continue Distributions

```python
def uniform_pdf(x):
    return 0 if x<= 0 else 1
```

```python
start = -10
end = 10
items = list(range(start, end))
values = [uniform_pdf(x) for x in items]

plt.title("PDF")
plt.bar(items, values)
plt.show()
```

```python
def uniform_cdf(x):
    if(x <= 0): return 0
    elif(x >= 1): return 1
    else: return x
```

```python
start = -1
end = 3
items = list(range(start, end))
values = [uniform_cdf(x) for x in items]

plt.title("CDF")
plt.plot(items, values)
plt.show()
```

# Normal Distribution

```python
def normal_distribution(x, mi, sigma):
    return (1/(math.sqrt(2 * math.pi)*sigma)) * math.exp(((-(x - mi)**2)) / 2*(sigma**2))
```

```python
numbers_range = range(-50, 50)
x_values_1 = [normal_distribution(x/4, mi=1, sigma=1) for x in numbers_range ]
x_values_2 = [normal_distribution(x/4, mi=1, sigma=0.5) for x in numbers_range ]
x_values_3 = [normal_distribution(x/4, mi=2, sigma=0.5) for x in numbers_range ]
x_values_4 = [normal_distribution(x/4, mi=2, sigma=1) for x in numbers_range ]

plt.title("Normal Distribution")
plt.plot(x_values_1, label="mi=1 sigma=1")
plt.plot(x_values_2, label="mi=1 sigma=0.5")
plt.plot(x_values_3, label="mi=2 sigma=0.5")
plt.plot(x_values_4, label="mi=2 sigma=1")
plt.legend()
plt.show()


```
